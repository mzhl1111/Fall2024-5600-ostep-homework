```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data extracted from the markdown
data = {
    'Implementation': [
        'cpp_thread_pool_100', 'cpp_thread_pool_boost', 'cpp_thread_pool_w_counter', 
        'go', 'cpp_async', 'rust_tokio', 'rust_thread_pool'
    ],
    'Requests/sec Avg': [
        6468.58, 6759.35, 6623.38, 449928.66, 6948.89, 139446.01, 90730.26
    ],
    'Requests/sec Stddev': [
        361.10, 366.04, 299.12, 3240, 417.88, 4890, 503.50
    ],
    'Requests/sec Max': [
        8350, 7900, 5290, 75130, 12200, 54150, 13210
    ],
    'Latency Avg (ms)': [
        10.09, 9.73, 10.13, 1.24, 9.65, 2.90, 1.96
    ],
    'Latency Stddev (ms)': [
        19.27, 16.65, 20.22, 1.13, 18.68, 2.81, 3.62
    ],
    'Latency Max (ms)': [
        300.85, 295.19, 335.55, 27.64, 315.98, 77.81, 208.19
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the style of the visualization
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(14, 10))
colors = sns.light_palette("green", as_cmap=True)

# Create a table heatmap for Requests/sec (Avg, Stddev, Max)
plt.subplot(2, 1, 1)
plot_requests = sns.heatmap(df[['Requests/sec Avg', 'Requests/sec Stddev', 'Requests/sec Max']].set_index(df['Implementation']), 
                            annot=True, fmt=".2f", cmap=colors, linewidths=.5, cbar=False)
plt.title('Performance Comparison (Requests/sec)')
plt.xlabel('')
plt.ylabel('')
```

```python
# Create a table heatmap for Latency (Avg, Stddev, Max)
plt.subplot(2, 1, 2)
plot_latency = sns.heatmap(df[['Latency Avg (ms)', 'Latency Stddev (ms)', 'Latency Max (ms)']].set_index(df['Implementation']), 
                           annot=True, fmt=".2f", cmap=colors, linewidths=.5, cbar=False)
plt.title('Latency Comparison (ms)')
plt.xlabel('')
plt.ylabel('')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
